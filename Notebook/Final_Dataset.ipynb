```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df1 = pd.read_csv('Received_Packets.csv')

df2 = pd.read_csv('collision_rate.csv')

df3 = pd.read_csv('Proba.csv')
```

```python
df1 =  df1[df1['node']!= 1]
df2 = df2[df2['node']!= 1]
df3 = df3[df3['node']!= 1]
```

```python
df1 = df1.set_index('node')
df2 = df2.set_index('node')
df3 = df3.set_index('node')
```

```python
df = pd.concat([df1,df2,df3],axis = 1)
```

```python

df.to_csv('FinalDataset.csv')
```
