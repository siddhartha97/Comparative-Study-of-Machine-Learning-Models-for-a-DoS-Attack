```python
import warnings
import numpy as np
import pandas as pd
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv('col1.txt',sep='delimiter',header=None)
df2 = pd.read_csv('col2.txt',sep='delimiter',header=None)
df3 = pd.read_csv('col3.txt',sep='delimiter',header=None)
df4 = pd.read_csv('col4.txt',sep='delimiter',header=None)
df5 = pd.read_csv('col5.txt',sep='delimiter',header=None)
```

```python
df1.columns = ["Packet Info"]
df2.columns = ["Time Stamp"]
df3.columns = ["Node"]
df4.columns = ["Trace Level"]
df5.columns = ["Column5"]
```

```python
frames = [df1,df2,df3,df4,df5]
```

```python
df = pd.concat(frames,axis=1)
```

```python
df.head()
```

```python
#Removed the unwanted stuffs from the following columns
df["Node"] = df["Node"].replace('_','',regex=True).astype(float)
```

```python
df.head(25)
```

# calculating collision rate

```python
filter_by_col = df[df['Column5']=='COL'].groupby('Node')
#col_rate = filter_by_col
time = filter_by_col['Time Stamp'].max() - filter_by_col['Time Stamp'].min()
collision_rate = filter_by_col.size()/time
collision_rate = pd.DataFrame(collision_rate,columns=['Collision_rate'])
```

```python
collision_rate = collision_rate.ix[1:,:]
collision_rate.to_csv('collision_rate.csv')
```

# calculating received rate

```python
received = df[df['Packet Info'] == 'r']
received = received[['Time Stamp','Node']]
group_node = df.groupby('Node')

received_packets = group_node.size()
time = group_node['Time Stamp'].max() - group_node['Time Stamp'].min()
received_rate = received_packets/time
```

```python
received_rate = pd.DataFrame(received_rate)
```

```python
received_rate = received_rate.rename(columns = {0:'Received Rate'})
```

```python
received_rate
```

```python
received_rate = received_rate.ix[1:,:]
received_rate.to_csv('Received_Packets.csv')
```
